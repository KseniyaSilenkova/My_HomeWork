## 1) Загрузите датасет с помощью функции np.genfromtxt()

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = np.genfromtxt(url, delimiter=',', dtype={'names':['sepal length', 'sepal width', 'petal length', 'petal width', 'species'], 'formats':['f8', 'f8', 'f8', 'f8', 'U20']})
```

```python
iris
```
